# 가상화폐 시세 판단하기

> 최고가와 최저가의 차이를 변동폭으로 정의할 때 (시가 + 변동폭)이 최고가 보다 높을 경우 "상승장", 그렇지 않은 경우 "하락장" 문자열을 출력하라.

|Key Name        |     Description|
|---|---|
|opeing_price    |최근 24시간 내 시작 거래금액|
|closing_price   |최근 24시간 내 마지막 거래금액|
|min_price       |최근 24시간 내 최저 거래금액|
|max_price       |최근 24시간 내 최고 거래금액|

````python
# 시가를 closing_price 로 잡는다
import requests
import pprint
btc = requests.get("https://api.bithumb.com/public/ticker/btc").json()['data']
pprint.pprint(btc)
````

````python
# 아래에 코드를 작성하세요.
flex = int(btc['max_price']) - int(btc['min_price'])
total = flex + int(btc['closing_price'])
if total > int(btc['max_price']):
    print('하락장')
else:
    print('사승장')
````

## 모든 가상화폐 시세 판단하기

> currrency dictionary에 담겨있는 모든 가상화폐 거래내역을 바탕으로 출력하라.
>
> 예외 처리에 유의하세요.
--- 
```
예시 출력)
BTC 상승장
ETH 상승장
DASH 상승장
LTC 하락장
ETC 하락장
XRP 상승장
BCH 상승장
...
```


````python
import pprint
currency = requests.get("https://api.bithumb.com/public/ticker/all").json()['data']
pprint.pprint(currency)
````

````python
# 아래에 코드를 작성하세요.
flexABT = float(currency['ABT']['max_price']) - float(currency['ABT']['min_price'])
resultABT = flexABT + float(currency['ABT']['closing_price'])
if resultABT > float(currency['ABT']['max_price']):
    print('상승장')
else:
    print('하락장')
````

````python
for coin in currency.keys():
    if 'max_price' in currency[coin]:
        flex = float(currency[coin]['max_price']) - float(currency[coin]['min_price'])
        result = flex + float(currency[coin]['closing_price'])
        if result > float(currency[coin]['max_price']):
            print(f'{coin} : 상승장')
        else:
            print(f'{coin} : 하락장')
````

````python
# 30% 이상이 상승장일때는 "곧 떡락합니다.", 50% 이상이 하락장일때는 "곧 떡상합니다." 출력하기
Uprising = 0
Downfalling = 0
for keys in currency.keys():
    if 'max_price' in currency[keys]:
        flex = float(currency[keys]['max_price']) - float(currency[keys]['min_price'])
        result = flex + float(currency[keys]['closing_price'])
        if result > float(currency[keys]['max_price']):
            print(f'{keys} : 상승장')
        else:
            print(f'{keys} : 하락장')
        if result > float(currency[keys]['max_price']):
            Uprising = Uprising + 1
        else:
            Downfalling = Downfalling + 1
            
if Uprising/(Uprising + Downfalling) > 0.3:
    print('곧 떡락합니다.')
if Downfalling/(Uprising + Downfalling) > 0.5:
    print('곧 떡상합니다.')
````

# 모음 제거하기

> 다음 문장의 모음을 제거하여 출력하세요.
---
```
예시 입력)
"Life is too short, you need python"
```

```
예시 출력)
Lf s t shrt, y nd pythn
```

````python
# 아래에 코드를 작성하세요.
ex = '\"Life is too short, you need python\"'
blank = ['']
answer = ''
for solving in ex:
    if not 'a' in solving:
            if not 'e' in solving:
                if not 'i' in solving:
                    if not 'o' in solving:
                        if not 'u' in solving:
                            blank.append(solving)
for result in blank:
    answer = answer + result
print(answer)

# 분명 좀 더 clear 한 방법이 있을거 같음
````

# 영어 이름 출력하기 

> 영어 이름은 가운데 이름을 가지고 있는 경우가 있습니다.
>
> 가운데 이름은 축약해서 나타내는 함수를 작성해보세요.

---
```
예시 입력)
Alice Betty Catherine Davis
```
```
예시 출력)
Alice B. C. Davis
```

````python
ex = 'Alice Betty Catherine Davis'
name = ''
list_ex = ex.split()
for nums in range(0,len(list_ex)):
    if not list_ex[0] == list_ex[-nums]:
        if not list_ex[-1] == list_ex[-nums]:
            list_ex[-nums] = list_ex[-nums][0:1] + '.'
for short_name in list_ex:
    name = name + short_name + ' '
print(name)
````
